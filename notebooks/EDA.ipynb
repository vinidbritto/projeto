```python
import pandas as pd
import plotly.express as px
import streamlit as st

df = pd.read_csv('../vehicles_us.csv')
```

```python
px.histogram(df, x='price', nbins=50, title='Distribuição de Preços')
```

```python
px.scatter(df, x='odometer', y='price', title='Preço vs Quilometragem')
```

```python
media_preco_tipo = df.groupby('type')['price'].mean().reset_index()
px.bar(media_preco_tipo, x='type', y='price', title='Preço Médio por Tipo de Carro')
```
